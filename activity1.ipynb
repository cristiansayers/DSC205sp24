QUESTION 2

Create a new column pop_density that shows the population density (inhabitants/sq mile) and output the states with the lowest and highest density.

```python
import pandas as pd
data = pd.read_csv('https://raw.githubusercontent.com/cristiansayers/DSC205sp24/main/USA%20State%20Data.csv')
data = data.drop(columns=[col for col in data if col.startswith('Unnamed')])
#First, the column values have to be converted from strings to floats
data['population_2022'] = data['population_2022'].str.replace(',','').astype(float)
data['area_sq_mile'] = data['area_sq_mile'].str.replace(',','').astype(float)

#Creating pop_density col
data['pop_density'] = data['population_2022'] / data['area_sq_mile']

#Finding the lowest and highest population density values
lowest_pop_dens = data.loc[data['pop_density'].idxmin()]
highest_pop_dens = data.loc[data['pop_density'].idxmax()]

#Printing the lowest and highest population density and their respective states
print(lowest_pop_dens['state_name'], "has the lowest population density with a value of",lowest_pop_dens['pop_density'])
print(highest_pop_dens['state_name'], "has the highest population density with a value of",highest_pop_dens['pop_density'])
```

QUESTION 3

Create a new column gdp_per_capita (gpd divided by population) and output the states with the lowest, average, and highest value.

```python
import pandas as pd
data = pd.read_csv('https://raw.githubusercontent.com/cristiansayers/DSC205sp24/main/USA%20State%20Data.csv')
data = data.drop(columns=[col for col in data if col.startswith('Unnamed')])
#Column values have to be converted from strings to floats
data['gdp_2022'] = data['gdp_2022'].str.replace(',','').astype(float)
data['population_2022'] = data['population_2022'].str.replace(',','').astype(float)

#Creating gdp_per_capita col
data['gdp_per_capita'] = data['gdp_2022'] / data['population_2022']

#Finding lowest, average, and highest GDP per capita values
lowest_gdp_percap = data.loc[data['gdp_per_capita'].idxmin()]
highest_gdp_percap = data.loc[data['gdp_per_capita'].idxmax()]

ave_gdp_percap = data['gdp_per_capita'].mean()

#Printing lowest, average, and highest GDP per capita values and their respective states (if applicable)
print(lowest_gdp_percap['state_name'], "has the lowest population density with a value of",lowest_gdp_percap['gdp_per_capita'])
print(highest_gdp_percap['state_name'], "has the highest population density with a value of",highest_gdp_percap['gdp_per_capita'])

print("The average GDP per capita is", ave_gdp_percap)
```

QUESTION 4

Show the states with gpd_per_capita that is greater than the average.  Only display the state abbreviation and the gdp_per_capita columns.
